```python
pd.pivot_table(data=transactions, columns="type", index="isFraud", values="amount", aggfunc="mean", fill_value=0)
```

Fraud only occurs for cash out and transfer transactions. It also tends to be higher amounts than the other types of transactions

```python
pd.pivot_table(data=transactions, index="isFraud", columns="isFlaggedFraud", values=["amount"], aggfunc="count")
```

998703 transactions were not flagged and were not fraudulent. 1296 were not flagged but were fraudulent. One was flagged and was fraudulent. None were flagged and not fraudulent.
